```r
library(dplyr)
library(lubridate)

prepare_day <- function(df) {
  df %>%
    mutate(
      Antrian = ymd_hms(Antrian),
      Dilayani = ymd_hms(Dilayani),
      Selesai = ymd_hms(Selesai)
    ) %>%
    arrange(Antrian) %>%
    filter(!is.na(Antrian), !is.na(Dilayani), !is.na(Selesai))
}

day1 <- prepare_day(day_1)
day2 <- prepare_day(day_2)
day3 <- prepare_day(day_3)
```

```r
compute_metrics <- function(df) {
  df %>%
    mutate(
      interarrival = c(NA, diff(Antrian)),
      service_time = Selesai - Dilayani,
      waiting_time = Dilayani - Antrian
    ) %>%
    mutate(across(c(interarrival, service_time, waiting_time),
                  ~ as.numeric(. , units = "secs")))
}

day1_m <- compute_metrics(day1)
day2_m <- compute_metrics(day2)
day3_m <- compute_metrics(day3)
```

```r
estimate_parameters <- function(df) {
  mean_interarrival <- mean(df$interarrival, na.rm = TRUE)
  mean_service <- mean(df$service_time, na.rm = TRUE)

  lambda <- 1 / mean_interarrival   # kedatangan per detik
  mu <- 1 / mean_service            # pelayanan per detik

  data.frame(
    mean_interarrival_sec = mean_interarrival,
    mean_service_sec = mean_service,
    lambda = lambda,
    mu = mu,
    rho = lambda / mu
  )
}

param_day1 <- estimate_parameters(day1_m)
param_day2 <- estimate_parameters(day2_m)
param_day3 <- estimate_parameters(day3_m)
```

```r
mm1_performance <- function(lambda, mu) {
  rho <- lambda / mu

  Lq <- rho^2 / (1 - rho)
  L <- rho / (1 - rho)
  Wq <- Lq / lambda
  W <- L / lambda

  data.frame(
    rho = rho,
    L = L,
    Lq = Lq,
    W = W,
    Wq = Wq
  )
}

perf_day1 <- mm1_performance(param_day1$lambda, param_day1$mu)
perf_day2 <- mm1_performance(param_day2$lambda, param_day2$mu)
perf_day3 <- mm1_performance(param_day3$lambda, param_day3$mu)
```

```r
library(readxl)
library(dplyr)
library(lubridate)
library(hms)

# --- Baca file ---
day_1 <- read_excel("data.xlsx", sheet = "Day 1")
day_2 <- read_excel("data.xlsx", sheet = "Day 2")
day_3 <- read_excel("data.xlsx", sheet = "Day 3")

day_1
day_2
day_3

# =========================================================
# 1. Fungsi bersihkan & hitung interarrival, service, waiting
# =========================================================

process_day <- function(df) {

  df <- df %>%
    arrange(Antrian) %>%
    mutate(
      interarrival = as.numeric(c(NA, diff(Antrian)), units = "secs"),
      service_time = as.numeric(Selesai - Dilayani, units = "secs"),
      waiting_time = as.numeric(Dilayani - Antrian, units = "secs")
    )

  # Buang baris tidak valid (NA, negatif)
  df_clean <- df %>%
    filter(
      !is.na(interarrival),
      !is.na(service_time),
      !is.na(waiting_time),
      interarrival >= 0,
      service_time > 0,
      waiting_time >= 0
    )

  return(df_clean)
}

clean_1 <- process_day(day_1)
clean_2 <- process_day(day_2)
clean_3 <- process_day(day_3)

clean_1
clean_2
clean_3

# =========================================================
# 2. Fungsi estimasi parameter M/M/1
# =========================================================

estimate_mm1 <- function(df_clean) {
  mean_interarrival <- mean(df_clean$interarrival)
  mean_service      <- mean(df_clean$service_time)

  lambda <- 1 / mean_interarrival
  mu     <- 1 / mean_service
  rho    <- lambda / mu

  return(
    data.frame(
      mean_interarrival_sec = mean_interarrival,
      mean_service_sec      = mean_service,
      lambda = lambda,
      mu = mu,
      rho = rho
    )
  )
}

mm1_1 <- estimate_mm1(clean_1)
mm1_2 <- estimate_mm1(clean_2)
mm1_3 <- estimate_mm1(clean_3)

mm1_1
mm1_2
mm1_3

# =========================================================
# 3. Fungsi hitung ukuran kinerja M/M/1
# =========================================================

mm1_metrics <- function(lambda, mu) {
  rho <- lambda / mu

  if (rho >= 1) {
    return(data.frame(rho = rho, L = NA, Lq = NA, W = NA, Wq = NA))
  }

  L  <- rho / (1 - rho)
  Lq <- rho^2 / (1 - rho)
  W  <- L / lambda
  Wq <- Lq / lambda

  return(data.frame(rho = rho, L = L, Lq = Lq, W = W, Wq = Wq))
}

perf_1 <- mm1_metrics(mm1_1$lambda, mm1_1$mu)
perf_2 <- mm1_metrics(mm1_2$lambda, mm1_2$mu)
perf_3 <- mm1_metrics(mm1_3$lambda, mm1_3$mu)
```

```r
# ----------------------------------------
# Membuat tabel-tabel laporan
# ----------------------------------------

library(dplyr)

# 1. Statistik Dasar
tabel_statistik <- bind_rows(
  clean_1 %>% mutate(Hari = "Day 1"),
  clean_2 %>% mutate(Hari = "Day 2"),
  clean_3 %>% mutate(Hari = "Day 3")
) %>%
  group_by(Hari) %>%
  summarise(
    jumlah_data = n(),
    mean_interarrival = mean(interarrival, na.rm = TRUE),
    mean_service = mean(service_time, na.rm = TRUE),
    min_interarrival = min(interarrival, na.rm = TRUE),
    max_interarrival = max(interarrival, na.rm = TRUE),
    min_service = min(service_time, na.rm = TRUE),
    max_service = max(service_time, na.rm = TRUE)
  )

# Memastikan final_params dan final_mm1 didefinisikan dari output 'Full CODE' cell
final_params <- bind_rows(
  mm1_1 %>% mutate(Hari = "Day 1"),
  mm1_2 %>% mutate(Hari = "Day 2"),
  mm1_3 %>% mutate(Hari = "Day 3")
)

final_mm1 <- bind_rows(
  perf_1 %>% mutate(Hari = "Day 1"),
  perf_2 %>% mutate(Hari = "Day 2"),
  perf_3 %>% mutate(Hari = "Day 3")
)

# 2. Parameter M/M/1
tabel_parameter <- final_params %>%
  select(Hari, lambda, mu, rho)

# 3. Karakteristik Sistem (M/M/1)
tabel_mm1 <- final_mm1 %>%
  select(Hari, rho, L, Lq, W, Wq)

# 4. Ringkasan Perbandingan
tabel_perbandingan <- final_params %>%
  mutate(
    Interpretasi = case_when(
      rho > 1 ~ "Sistem tidak stabil",
      rho < 0 ~ "Data tidak valid (service > arrival secara tidak wajar)",
      TRUE ~ "Sistem stabil"
    )
  )

# 5. Tabel Stabilitas
tabel_stabilitas <- final_params %>%
  mutate(
    Keterangan = ifelse(rho < 1, "Stabil", "Tidak Stabil")
  ) %>%
  select(Hari, rho, Keterangan)
```

```r
mm2_performance <- function(lambda, mu) {
  rho <- lambda / (2 * mu)

  if (rho >= 1) {
    return(data.frame(rho = rho, L = NA, Lq = NA, W = NA, Wq = NA))
  }

  # Probability of 0 customers in the system
  P0 <- 1 / ( (1) + (lambda / mu) + ( (lambda / mu)^2 / (2 * (1 - rho)) ) )

  # Lq (Expected number of customers in the queue)
  Lq <- (P0 * (lambda / mu)^2 * rho) / (2 * (1 - rho)^2)

  # L (Expected number of customers in the system)
  L <- Lq + (lambda / mu)

  # Wq (Expected waiting time in the queue)
  Wq <- Lq / lambda

  # W (Expected waiting time in the system)
  W <- Wq + (1 / mu)

  return(data.frame(rho = rho, L = L, Lq = Lq, W = W, Wq = Wq))
}
```

```r
mm2_perf_1 <- mm2_performance(mm1_1$lambda, mm1_1$mu)
mm2_perf_2 <- mm2_performance(mm1_2$lambda, mm1_2$mu)
mm2_perf_3 <- mm2_performance(mm1_3$lambda, mm1_3$mu)

mm2_perf_1
mm2_perf_2
mm2_perf_3
```

```r
mm3_performance <- function(lambda, mu) {
  c <- 3 # Number of servers
  rho <- lambda / (c * mu)

  if (rho >= 1) {
    return(data.frame(rho = rho, L = NA, Lq = NA, W = NA, Wq = NA))
  }

  # Calculate P0 (Probability of 0 customers in the system)
  # Summation part: (lambda/mu)^n / n! for n from 0 to c-1
  sum_part <- sum(sapply(0:(c-1), function(n) {
    if (n == 0) 1
    else (lambda/mu)^n / factorial(n)
  }))

  # Last term: ((lambda/mu)^c / c!) * (1 / (1 - rho))
  last_term <- ((lambda/mu)^c / factorial(c)) * (1 / (1 - rho))

  P0 <- 1 / (sum_part + last_term)

  # Lq (Expected number of customers in the queue)
  Lq <- ( (lambda/mu)^c * rho ) / ( factorial(c) * (1 - rho)^2 ) * P0

  # L (Expected number of customers in the system)
  L <- Lq + (lambda / mu)

  # Wq (Expected waiting time in the queue)
  Wq <- Lq / lambda

  # W (Expected waiting time in the system)
  W <- Wq + (1 / mu)

  return(data.frame(rho = rho, L = L, Lq = Lq, W = W, Wq = Wq))
}
```
